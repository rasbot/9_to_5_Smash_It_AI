```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.style.use('fivethirtyeight')
plt.style.use('dark_background')
plt.style.use('seaborn-bright')
```

```python
def get_reward_data(file_path):
    df = pd.read_csv(file_path)
    df = df.dropna(axis=0)
    return df['Steps'], df['Policy/Extrinsic Reward']
```

```python
# Comparing training times between PPO and SAC Policies.
# Get timing data from csvs
sac_times = pd.read_csv('sac_times1.csv',header=None)[2]
ppo_times = pd.read_csv('ppo_times.csv',header=None)[2]
ppo_3agents_times = pd.read_csv('time_3agents.csv',header=None)[2]
ppo_6agents_times = pd.read_csv('time_6agents.csv',header=None)[2]
ppo_9agents_times = pd.read_csv('time_9agents.csv',header=None)[2]
ppo_30agents_times = pd.read_csv('time_30agents.csv',header=None)[2]
```

```python
df1_x, df1_y = get_reward_data('results/walker_T1/Walker.csv')
df4_x, df4_y = get_reward_data('results/walker_T4/Walker.csv')
df5_x, df5_y = get_reward_data('results/walker_T5/Walker.csv')
df6_x, df6_y = get_reward_data('results/walker_T6/Walker.csv')
df7_x, df7_y = get_reward_data('results/walker_T7/Walker.csv')
df8_x, df8_y = get_reward_data('results/walker_T8/Walker.csv')
df9_x, df9_y = get_reward_data('results/walker_T9/Walker.csv')
df10_x, df10_y = get_reward_data('results/walker_T10/Walker.csv')
df11_x, df11_y = get_reward_data('results/walker_T11/Walker.csv')
df12_x, df12_y = get_reward_data('results/walker_T12/Walker.csv')
df13_x, df13_y = get_reward_data('results/walker_T13/Walker.csv')
df14_x, df14_y = get_reward_data('results/walker_T14/Walker.csv')
df15_x, df15_y = get_reward_data('results/walker_T15/Walker.csv')
df16_x, df16_y = get_reward_data('results/walker_SAC1/Walker.csv')
df17_x, df17_y = get_reward_data('results/walker_3bois/Walker.csv')
df18_x, df18_y = get_reward_data('results/walker_6bois/Walker.csv')
df19_x, df19_y = get_reward_data('results/walker_9bois/Walker.csv')
df20_x, df20_y = get_reward_data('results/walker_30bois/Walker.csv')
df21_x, df21_y = get_reward_data('results/walker_30boi_long/Walker.csv')
df22_x, df22_y = get_reward_data('results/walker_lanky/Walker.csv')
```

```python
fig, axs = plt.subplots(figsize=(12,10))
axs.plot(df1_x.iloc[:669], df1_y.iloc[:669], label='default parameters')
axs.plot(df21_x.iloc[:200], df21_y.iloc[:200], label='tuned parameters')
axs.grid(False)
axs.legend(fontsize=20)
axs.set_xlabel('Step Number')
axs.set_ylabel('Policy Reward')
axs.set_title('Policy Reward vs Step for 2 Runs')
plt.savefig('reward_2runs_trunc.png')
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12,10))
axs.plot(df1_x, df1_y, label='default parameters')
axs.plot(df21_x, df21_y, label='tuned parameters')
axs.grid(False)
axs.legend(fontsize=20)
axs.set_xlabel('Step Number')
axs.set_ylabel('Policy Reward')
axs.set_title('Policy Reward vs Step for 2 Runs')
plt.savefig('reward_2runs.png')
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12,10))
axs.plot(df6_x, df6_y, label='1 layer')
axs.plot(df4_x, df4_y, label='2 layer')
axs.plot(df5_x, df5_y, label='3 layer')
axs.plot(df7_x, df7_y, label='4 layer')
axs.grid(False)
axs.legend(fontsize=15)
axs.set_xlabel('Step Number')
axs.set_ylabel('Policy Reward')
axs.set_title('Policy Reward Progression for Different Number of Hidden Layers')
plt.savefig('reward_layers.png')
plt.show()
```
